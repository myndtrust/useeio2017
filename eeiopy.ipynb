# EEIO Program

## Economic Input/Output Tables and Activities


<br>**I_matrix** {{I_matrix.shape}}
<br>Identity Matrix
<br>$I$

**Matrix_A** {{Matrix_A.shape}}
<br>Technology matrix, also called the direct requirements matrix. The form is commodity input (rows) per commodity output (columns). The technology matrix is obtained from EPA USEEIO 2007 dataset. And it is also complied with the 2007 Input-Output (IO) tables published by the US Bureau of Economic Analysis (BEA 2015). This is calculated in the IO Model Builder by multiplying the commodity-by-industry direct requirements derived from the Use table by the market share matrix (industry-by-commodity) derived from the Make table after the scrap adjustment procedure.
<br>direct  requirements  table(or  matrix)

```python
 Matrix_A.head()
```

## IplusA
<br>Adding the direct requirement matrix to the identity matrix indicates the direct share of requirements to the index's industrial sector.

```python
IplusA = I_matrix + Matrix_A
IplusA.head()
```

```python
IplusA_total =pd.DataFrame(IplusA.sum(axis=(1)))
IplusA_total.columns = ['Total']
IplusA_total.head()
```

## IminusA
<br>Subtracting the direct requirements matrix from the identity matrix indicates the total share of requirements to the index's industrial sector.

```python
IminusA=np.subtract(I_matrix,Matrix_A)
IminusA.head()
```

## inv_IminusA
<br>This is the Leontif inverse of $A$. 

*Leontief_inverse_of_A* 
<br>$[I-A]^{-1}$

```python
inv_IminusA = pd.DataFrame(np.linalg.inv(IminusA))
inv_IminusA.index = IminusA.index
inv_IminusA.columns = IminusA.columns
inv_IminusA.head()
```

```python
IminusA_total =pd.DataFrame(IminusA.sum(axis=(1)))
IminusA_total.columns = ['Total']
IminusA_total.head()
```

<br>**TRACI, what?**, The U.S. Environmental Protection Agency developed an Impact Assessment methodology called TRACI, short for "Tool for the Reduction and Assessment of Chemical and Other Environmental Impacts

**Tool for Reduction and Assessment of Chemicals and Other Environmental Impacts (TRACI)**

Responsible decision-making seeks to minimize negative impacts while balancing environmental, economic, and social factors.

TRACI is an environmental impact assessment tool. It provides characterization factors for Life Cycle Impact Assessment (LCIA), industrial ecology, and sustainability metrics. Characterization factors quantify the potential impacts that inputs and releases have on specific impact categories in common equivalence units. Impact categories include: 

ozone depletion,
climate change,
acidification,
eutrophication,
smog formation,
human health impacts, and
ecotoxicity.
Resource uses of fossil fuels are also characterized.

Users of TRACI include LCIA practitioners and researchers. TRACI can be applied to processes, products, facilities, companies and communities. TRACI can be updated as advances in LCIA research occur.
